```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

## Ejercicio 1

Un perceptrón o neurona no es capaz de aproximar una función que separe en dos clases (naranja y azul) el conjunto de datos mostrado en la figura 1. Propón una transformación aplicada a los atributos $x_{1}$ y $x_{2}$ de forma que puedan ser separados por solo un perceptrón. Justifica porque la transformación propuesta funciona. (2 puntos)

<img src="https://raw.githubusercontent.com/davidguzmanr/Aprendizaje-Profundo/main/Tareas/Images/tarea1-1.png" width=300 height=300 />

**Solución**. Pasemos a coordenadas polares, es decir, apliquemos la transformación

$$ (x_{1}, x_{2}) \longrightarrow \left( \sqrt{x_{1}^{2} + x_{2}^{2}}, \ \tan^{-1}\left( \frac{x_{2}}{x_{1}} \right) \right). $$

Como se observa en la figura de abajo con esta transformación ahora los datos son fácilmente separables con una simple línea, de modo que pueden ser separados con sólo un perceptrón.

```python
X, y = make_circles(n_samples=1000, factor=0.1, noise=0.05)
X_transformed = np.array([[np.sqrt(x[0]**2 + x[1]**2), np.arctan(x[1]/x[0])] for x in X])

colors = ['#F2A30E' if i == 1 else '#007AEC' for i in y]
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

ax0.scatter(X[:,0], X[:,1], c=colors, s=2)
ax0.set_title('Datos originales', weight='bold')

ax1.scatter(X_transformed[:,0], X_transformed[:,1], c=colors, s=2)
ax1.set_title('Datos transformados', weight='bold')

plt.show()
```
